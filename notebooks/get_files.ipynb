```python
# -*- coding: utf-8 -*-
"""
Created on Oct 2018
"""

import concurrent
import gcsfs
import os
import pandas as pd
import shutil
 
from gcloud import storage
 
class ReadBucket(object):
    """Module to read data from bucket"""
 
    def __init__(self, bucket_name, project_name):
        """Init.
 
        :param bucket_name: (str) name of the bucket
        """
        self.bucket_name = bucket_name
        self.client = storage.Client()
        self.project_name = project_name
        self.fs = gcsfs.GCSFileSystem(project=self.project_name)
        self.bucket = self.client.get_bucket(self.bucket_name)
 
    def list_files_bucket(self, directory=None, delimiter=None):
        """List all files in the bucket
 
        :param directory: (str) directory path/to/data
        :param delimiter: (str) restrict the results to only the "files" in the given "folder"
        :return: (list) data files
        """
        blobs = self.bucket.list_blobs(prefix=directory, delimiter=delimiter)
 
        return list(blobs)
 
    def read_blob_as_pandas(self, blob, encoding="utf-8", sep=","):
        """Read a blob as pandas Dataframe
 
        :param blob: (Blob) datapath
        :param encoding: (str) encoding
        :param sep: (str) separator
        :return: (pd.DataFrame) dataset
        """
        file = blob.bucket.name + '/' + blob.name
        with self.fs.open(file) as f:
            df = pd.read_csv(f, encoding=encoding, sep=sep)
        print(df.shape)
 
        return df

    def copying_blob_files(self, blob, destination_dir):
        source_file = blob.bucket.name + '/' + blob.name
        destination_file = destination_dir + '/' + blob.name
        path = "/".join(destination_file.split("/")[:-1])
        os.makedirs(path, exist_ok=True)
        print(source_file, destination_file)
        with self.fs.open(source_file, 'rb') as source_file:
            with open(destination_file, 'wb+') as destination_file:
                shutil.copyfileobj(source_file, destination_file)
                
    def copying_all_blob_files(self, blob_list, destination_dir):
        with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
            executor.map(lambda blob: self.copying_blob_files(blob, destination_dir), blob_list)
```

```python
rb = ReadBucket("comp-car-storage", "mood-detection-poc")
file_list = rb.list_files_bucket()
# creer un folder data avant
rb.copying_all_blob_files(file_list, "data")
```
